```python
# Для реалізації нижче наведених функцій використовував ресурси з https://habr.com/ru/articles/212235/ для реалізації системи AES-128 
# Але переробив під умови завдання (роботу з трьома системами - AES-128, AES-192 i AES-256)

def sub_bytes(state, inv=False):
    """
    Перетворення є заміною кожного байта з State на відповідний йому з константної таблиці InvSbox.
    """
    if not inv: # Шифрування
        box = sbox
    else:   # Дешифрування
        box = inv_sbox
    
    for i in range(len(state)):
        for j in range(len(state[i])):
            row = state[i][j] // 0x10
            col = state[i][j] % 0x10

            box_elem = box[16*row + col]
            state[i][j] = box_elem
    return state

def shift_rows(state, inv=False):
    """
    Трансформація здійснює циклічний зсув вправо на 1 елемент для першого рядка State, на 2 для другого та на 3 для третього. Нульовий рядок не повертається.
    """
    if not inv: # Шифрування
        count = 1
        for i in range(1, 4):
            state[i] =  state[i][count:count+4] +  state[i][0:count]
            count += 1
    else: # Дешифрування
        count = 3
        for i in range(1, 4):
            state[i] =  state[i][count:count+4] + state[i][0:count]
            count -= 1
    return state

"""
У рамках цієї трансформації кожна колонка в State представляється у вигляді многочлена та перемножується в полі GF(2^8) за модулем x^4 + 1 з фіксованим багаточленом {0b}x^3 + {0d}x^2 + {09}x + {0e}.
"""

# Допоміжні функції множення

def mul_by_02(num):
    if num < 0x80:
        res = (num << 1)
    else:
        res = (num << 1)^0x1b

    return res % 0x100

def mul_by_03(num):
    return mul_by_02(num)^num

def mul_by_09(num):
    return mul_by_02(mul_by_02(mul_by_02(num)))^num
 
def mul_by_0b(num):
    return mul_by_02(mul_by_02(mul_by_02(num)))^mul_by_02(num)^num

def mul_by_0d(num):
    return mul_by_02(mul_by_02(mul_by_02(num)))^mul_by_02(mul_by_02(num))^num

def mul_by_0e(num):
    return mul_by_02(mul_by_02(mul_by_02(num)))^mul_by_02(mul_by_02(num))^mul_by_02(num)

# Власне трансформація

def mix_columns(state, inv=False):
    for i in range(4):
        if not inv: # Шифрування
            s0 = mul_by_02(state[0][i])^mul_by_03(state[1][i])^state[2][i]^state[3][i]
            s1 = state[0][i]^mul_by_02(state[1][i])^mul_by_03(state[2][i])^state[3][i]
            s2 = state[0][i]^state[1][i]^mul_by_02(state[2][i])^mul_by_03(state[3][i])
            s3 = mul_by_03(state[0][i])^state[1][i]^state[2][i]^mul_by_02(state[3][i])
        else: # Дешифрування
            s0 = mul_by_0e(state[0][i])^mul_by_0b(state[1][i])^mul_by_0d(state[2][i])^mul_by_09(state[3][i])
            s1 = mul_by_09(state[0][i])^mul_by_0e(state[1][i])^mul_by_0b(state[2][i])^mul_by_0d(state[3][i])
            s2 = mul_by_0d(state[0][i])^mul_by_09(state[1][i])^mul_by_0e(state[2][i])^mul_by_0b(state[3][i])
            s3 = mul_by_0b(state[0][i])^mul_by_0d(state[1][i])^mul_by_09(state[2][i])^mul_by_0e(state[3][i])

        state[0][i] = s0
        state[1][i] = s1
        state[2][i] = s2
        state[3][i] = s3
 
    return state 

"""
Ця допоміжна трансформація формує набір раундових ключів – KeySchedule. 
KeySchedule являє собою довгу таблицю, що складається з Nb * (Nr + 1) стовпців або (Nr + 1) блоків, 
кожен з яких дорівнює розміру State. Перший раундовий ключ заповнюється на основі секретного ключа, який ви придумаєте, 
за формулою KeySchedule[r][c] = SecretKey[r + 4c], r = 0,1...4; c = 0,1 ... Nk.
"""

def key_expansion(key, Nk, Nb, Nr):
    key_symbols = [symbol for symbol in key]

    key_schedule = [[] for i in range(4)]     
    for r in range(4):
        for c in range(Nk):
            key_schedule[r].append(key_symbols[r + 4*c])
    
    for col in range(Nk, Nb*(Nr + 1)): # col - column number
        tmp = [key_schedule[row][col-1] for row in range(0, 4)]
        if col % Nk == 0:
            #rotate
            tmp = [key_schedule[row][col-1] for row in range(1, 4)]
            tmp.append(key_schedule[0][col-1]) 
            
            # change its elements using Sbox-table like in SubBytes
            for j in range(len(tmp)):        
                sbox_row = tmp[j] // 0x10
                sbox_col = tmp[j] % 0x10
                sbox_elem =  sbox[16*sbox_row + sbox_col]
                tmp[j] = sbox_elem
            
            tmp[0] ^= rcon[col//Nk]
                
        elif Nk > 6 and col % Nk == 4:
            for j in range(len(tmp)):        
                sbox_row = tmp[j] // 0x10
                sbox_col = tmp[j] % 0x10
                sbox_elem =  sbox[16*sbox_row + sbox_col]
                tmp[j] = sbox_elem
                
        for row in range(4):
            s = key_schedule[row][col - Nk]^tmp[row]
            key_schedule[row].append(s)
            
    return key_schedule
            
"""
Трансформація виробляє побітовий XOR кожного елемента з State з відповідним елементом з RoundKey. 
RoundKey - масив такого ж розміру, як State, який будується для кожного раунду на основі секретного ключа функцією KeyExpansion()
"""

def add_round_key(state, key_schedule, Nk, Nb, round=0):
    for col in range(4):
        # Nb*round is a shift which indicates start of a part of the KeySchedule
        s0 = state[0][col]^key_schedule[0][Nb*round + col]
        s1 = state[1][col]^key_schedule[1][Nb*round + col]
        s2 = state[2][col]^key_schedule[2][Nb*round + col]
        s3 = state[3][col]^key_schedule[3][Nb*round + col]

        state[0][col] = s0
        state[1][col] = s1
        state[2][col] = s2
        state[3][col] = s3

    return state


def encryption(input_bytes, key, size):
    Nb = 4  # Кількість стовпчиків у State
    Nk = size // 4 
    Nr = 0  # Кількість раундів шифрування

    size_to_nr = {16: 10, 24: 12, 32: 14}  # Створення словника, щоб програма працювала для декількох розмірів
    Nr = size_to_nr.get(size)
    if Nr is None:
        return None


    state = [[] for j in range(Nb)]
    for r in range(4):
        for c in range(Nb):
            state[r].append(input_bytes[r + 4*c])

    key_schedule = key_expansion(key, Nk, Nb, Nr)  # Розширення ключа
    state = add_round_key(state, key_schedule, Nk, Nb)  # Додавання ключа раунду до State

    for rnd in range(1, Nr):
        state = sub_bytes(state)  # Заміна байтів у State по таблиці
        state = shift_rows(state)  # Зсув рядків State
        state = mix_columns(state)  # Множення стовпців State на фіксовану матрицю
        state = add_round_key(state, key_schedule, Nk, Nb, rnd)  # Додавання ключа раунду до State

    state = sub_bytes(state)  # Заміна байтів у State по таблиці
    state = shift_rows(state)  # Зсув рядків State
    state = add_round_key(state, key_schedule, Nk, Nb, rnd + 1)  # Додавання ключа останнього раунду до State

    return [state[r][c] for c in range(Nb) for r in range(4)] # Повернення State у вихідний блок



def decryption(cipher, key, size):
    Nb = 4  # Кількість стовпчиків у State
    Nk = size // 4 
    Nr = 0  # Кількість раундів розшифрування

    size_to_nr = {16: 10, 24: 12, 32: 14}  # Створення словника, щоб програма працювала для декількох розмірів
    Nr = size_to_nr.get(size)
    if Nr is None:
        return None

    state = [[] for i in range(Nb)]
    for r in range(4):
        for c in range(Nb):
            state[r].append(cipher[r + 4*c])

    key_schedule = key_expansion(key, Nk, Nb, Nr)  # Розширення ключа
    
    state = add_round_key(state, key_schedule, Nk, Nb, Nr)  # Додавання ключа останнього раунду до State

    rnd = Nr - 1
    while rnd >= 1:
        state = shift_rows(state, inv=True)  # Зсув рядків State у зворотньому напрямку
        state = sub_bytes(state, inv=True)  # Заміна байтів у State зворотньою заміною по таблиці
        state = add_round_key(state, key_schedule, Nk, Nb, rnd)  # Додавання ключа раунду до State
        state = mix_columns(state, inv=True)  # Зворотнє множення стовпців State на фіксовану матрицю
        
        rnd -= 1

    state = shift_rows(state, inv=True)  # Зсув рядків State у зворотньому напрямку
    state = sub_bytes(state, inv=True)  # Заміна байтів у State зворотньою заміною по таблиці
    state = add_round_key(state, key_schedule, Nk, Nb, rnd)  # Додавання ключа першого раунду до State

    return [state[r][c] for c in range(Nb) for r in range(4)]  # Повернення State у вихідний блок
```

### Тестові значення для перевірки взято з підручника

```python
# Тест для AES-128

plaintext = "00112233445566778899aabbccddeeff"
key = "000102030405060708090a0b0c0d0e0f"

# Перетворення з рядка в байти
plaintext = bytes.fromhex(plaintext) 
key = bytes.fromhex(key) 

# Шифрування
byte_list_e = encryption(plaintext, key, 16)

# Розшифрування
byte_list_d = decryption(byte_list_e, key, 16)

# Перетворення з байтів в рядок
ciphertext = ''.join([format(byte, '02x') for byte in byte_list_e])
decrypted = ''.join([format(byte, '02x') for byte in byte_list_d])

print("Ciphertext for AES-128:", ciphertext)
print("Decrypted text for AES-128: ", decrypted)
```
